Suite de Fibo

```python
# Suite de fibo 

def get_fibo(n):
    a,b = 0,1
    c = b
    sequence = [b]
    if n in range(1):
        return(b)
    for i in range (1,n):
        a,b = b,c 
        c = a + b
        sequence.append(b)
    return sequence

get_fibo(10)

    
```
